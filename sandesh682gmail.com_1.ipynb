```python
# So let us plot 1D plot first:

import numpy as np
haberman_1 = haberman.loc[haberman["Surv_status"] == 1];
haberman_2 = haberman.loc[haberman["Surv_status"] == 2];


plt.plot(haberman_1["Age"], np.zeros_like(haberman_1['Age']), 'o',label='1')
plt.plot(haberman_2["Age"], np.zeros_like(haberman_2['Age']), 'o',label='2')
plt.legend(title = "Surv_status")
plt.title("1D plot with Featue Age")
```

```python
# Let us find the best feature or column for plotting 1D plots using Univarient analysis:
#Checking For Age

sns.FacetGrid(haberman, hue="Surv_status", size=5) \
   .map(sns.distplot, "Age") \
   .add_legend();
plt.title("Histogram for feature Age")
plt.show();
```

```python
#Let us plot CDF for checking out probabilities with diffrent features:
#CDF for Age:

counts, bin_edges = np.histogram(haberman_1['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf,label = "1")


counts, bin_edges = np.histogram(haberman_1['Age'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.xlabel("Age")
plt.title("CDF for patients who will surivive more than 5 years:")
plt.ylabel("Probability")
plt.legend()
plt.show();
```

Observations:

1.There are 75% patients who are going to survive more than 5 years whose age is less than 60

```python

counts, bin_edges = np.histogram(haberman_1['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf,label = "1")

counts, bin_edges = np.histogram(haberman_2['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf,label = "2")

plt.xlabel("Age")
plt.title("CDF for patients who will surivive more than 5 years and less than 5 years:")
plt.ylabel("Probability")
plt.legend()
```

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(haberman_1["Axil_nodes"]))
#Mean with an outlier.
print(np.mean(np.append(haberman_1["Axil_nodes"],50)));
print(np.mean(haberman_2["Axil_nodes"]))

print("\nStd-dev:");
print(np.std(haberman_1["Axil_nodes"]))
print(np.std(haberman_2["Axil_nodes"]))
```

Observations:

1.The mean value for count of Axial_nodes for patients who will survive more than 5 years is 2.79

2.The mean value for count of Axial_nodes for patients who will survive less than 5 years is 7.45

3.Hence axial_nodes is a better feature to classify the survival_status

4.Standard deviation values are more and hence we cannot depend only on Axil_nodes for classification

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(haberman["Axil_nodes"]))
#Median with an outlier
print(np.median(np.append(haberman["Axil_nodes"],50)));
print(np.median(haberman_2["Axil_nodes"]))

print("\nQuantiles:")
print(np.percentile(haberman["Axil_nodes"],np.arange(0, 100, 25)))
print(np.percentile(haberman_2["Axil_nodes"],np.arange(0, 100, 25)))


print("\n90th Percentiles:")
print(np.percentile(haberman_1["Axil_nodes"],90))
print(np.percentile(haberman_2["Axil_nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(haberman["Axil_nodes"]))
print(robust.mad(haberman_2["Axil_nodes"]))
```

Observations:

1.Median values does not give any clear idea to make any clear conclusion

2.90th percentile values of Axil_nodes give basic ideas about Surv_status of patient
